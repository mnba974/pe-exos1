```python
def cadre(a):
    m=np.zeros((1,len(a[0])))
    a[1]=m
    a[-2]=m
    n=np.zeros((1,len(a)))
    a[:,1]=n
    a[:,-2]=n
cadre(a)
plt.imshow(a)
```

```python
b=np.random.randint(20,size=(6,8))
cadre(b)
plt.imshow(b)
```

```python
def cadre2(a):
    m=np.zeros((1,len(a)))
    a=np.insert(a,0,m,1)
    a=np.insert(a,len(a[0]),m,1)
    n=np.zeros((1,len(a[0])))
    a=np.insert(a,0,n,0)
    a=np.insert(a,len(a),n,0)
    return a
b=cadre2(b)
plt.imshow(b)
```

```python
def anti_diagonal(l):
    n=len(l)
    r=np.zeros((n,n))
    for i in range(n):
        r[i][n-i-1]=l[i]
    return r
print(anti_diagonal([1,2,3,4]))
print(anti_diagonal([1]))
print(anti_diagonal([1,2]))
```

```python
def anti_diagonal2(l):
    n=len(l)
    r=np.empty((n,n))
    for i in range(n):
        r[i][n-i-1]=l[i]
    return r
print(anti_diagonal2([1,2]))
```
